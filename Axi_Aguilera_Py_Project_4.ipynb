# **Analytical Analysis**

As we know $x$ is a critical point of $f(x)$ if and only if $f'(x) =0$. Then computing the $f'$ and applying the criterion, we obtain

$$f'(x) = \dfrac{x}{\sqrt{x^2 + 5}} = 0 \qquad ⟺ x = 0$$

Now, note that     $f''(x) = \dfrac{5}{(x^2 + 5)^{3/2}}$  and $f''(0) = 5^{-1/2}>0$.

Hence $f(x)$ has a minimum at $x = 0$ and the minimun value of $f$ at this point is  $f(0) = \sqrt{5}$.

# **Numerical Analysis**

Now, we compute the minimum value of $f(x) =\sqrt{x^2 + 5}$ on $[-5,5]$ using a numerical approach.

First, we define the function $ f(x)$, then constructs a grid of $1000$ evenly spaced points across the interval using `np.linspace`. Then we evaluate $ f(x) $ at these points and plots the resulting curve using Matplotlib.

The graph highlights the minimum point, located at $ x = 0 $, with a dashed vertical line and a red marker for minimum value, $ \sqrt{5}$.

```python
import numpy as np


# Define the function f(x).
def f(x):
  return np.sqrt(x**2 + 5)

# First we construct a grid of the interval [-5,5] and compute the values of f
x_values = np.linspace(-5,5,1000)
f_values = f(x_values)
```

```python

import matplotlib.pyplot as plt

# Plot of f(x) using x_values and the f_values
plt.figure(figsize=(8, 8))
plt.plot(x_values, f_values, label=r"$f(x) = \sqrt{x^2 + 5}$", color='blue')
plt.axvline(0, color='black', linestyle='--', alpha=0.7, label="Minimum at $x = 0$")
plt.scatter(0, f(0), color='red', label=r"Minimum $f(0) = \sqrt{5}$")
plt.title("Plot of $f(x)$ over $[-5, 5]$")
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.legend(loc='best')
plt.grid()
plt.show()

```

To compute the minimun value of $ f(x) = \sqrt{x^2 + 5} $ we implement a famous optimization algorithm, Gradient Descent.

Gradient Descent is commonly used to minimize functions by iteratively updating the values of $ x $ in the direction of the negative gradient (the steepest descent).

**Steps in the Code**:
1. **Derivative of \( f(x) \)**:
   The first part of the code defines the derivative of $ f(x) $, which is needed for the gradient descent updates.

2. **Gradient Descent Function**:
   A function `gradientDescent` is defined to run the algorithm. It takes as input the initial value $ x_0 $, the step size $ \alpha $, and the number of iterations $ T $. At each iteration, the algorithm updates $ x_k $ using the recursive relation:

   $$ x_{k+1} = x_k - \alpha f'(x_k) $$

   This formula moves $ x $ in the direction of the negative gradient, which should bring us closer to the minimum of $ f(x) $.

3. **Initialization**:
   The initial value $ x_0 $ is set based on the student's ID number. The step size $ \alpha $ is set to $1$, and the number of iterations $ T $ is set to $50$.

4. **Running the Algorithm**:
   The gradient descent function is then called with the initialized values. After 50 iterations, the final value $ x_T $ is the point where the function reaches its minimum, and the corresponding value of $ f(x_T) $ is printed.

5. **Plotting the Results**:
   The plot visualizes how the function value $f(x_k) $ changes with each iteration. The red curve shows the value of $ f(x_k) $ over the iterations, while the blue line represents the minimum value of the function, which occurs at $ x = 0 $ and is equal to $ \sqrt{5} $.


```python
# Define the derivative of the function f(x), for gradient descent
def derivative_f(x):
  return x/np.sqrt(x**2 + 5)

# write a function to run the Gradient Descent Algorithm, take input arguments x_0, step_size, and num_iterations; return a list/array of x_k values.
def gradientDescent(x_0, step_size, num_Iter):
  x_min = [x_0]

  for k in range(1,num_Iter + 1):
    x_current = x_min[-1] - step_size*derivative_f(x_min[-1])
    x_min.append(x_current)

  return np.array(x_min)


# set up/assign initial values
ID_number = 116480365
x_0 = ID_number % 10
alpha = 1
T = 50


# Run gradient descent by calling the function
x_min = gradientDescent(x_0 = x_0, step_size = alpha, num_Iter = T)

# report x_T by printing out the value
x_T = x_min[-1]
print(f"After  50  iterations the function f(x) reach its mininum value at x = {x_T}")

# Plot the f(x_k) values vs. iteration number
y = 0
plt.figure(figsize=(8, 8))
plt.plot(range(T + 1), f(x_min), label=r"$f(x_k)$ over iterations", color='red', marker='x', markeredgewidth=2)
plt.axhline(y=np.sqrt(5),  label='$f(0) = \sqrt{5}$', color='blue', linestyle='-', linewidth=2)
plt.title("Value of $f(x_k)$ vs Number of Iteration $(k)$")
plt.xlabel("Iteration $k$")
plt.ylabel("$f(x_k)$")
plt.legend(loc='best')
plt.grid()
plt.show()

```

Using a loop we iterate through each step size $\alpha$, calling the `gradientDescent` function to perform the gradient descent and find the minimum of the function. The results are stored in the results dictionary with $\alpha$ as key.

For each step size, the difference between the function values at consecutive iterations is computed as $$f(x_{k-1}) - f(x_k)$$ and plotted against the iteration number $k$. The plot shows how the difference in function values decreases over time, indicating convergence. Smaller differences as $k$ increases suggest that the algorithm is approaching a minimum.

In this section, we aim to find the coefficients (weights) of a linear equation that best fit a $2D$ data.

1. **Data Generation:**
  A set of $100$ two-dimensional data points $ X = [x_1, x_2] $ is created. The target values $y$ are generated using the linear equation:

  $$ y = 4 + 4x_1 + 0.5x_2 + \epsilon $$

  where $ \epsilon $ is random noise introduced to make the data more realistic. The goal of linear regression is to find the best-fitting line (or plane in this case) through the data.

```python
import numpy as np

# Generate synthetic data
np.random.seed(65)   # change the seed to be the last two digits of your SB ID.
X = 2 * np.random.rand(100, 2)  # Generate 100 random 2D data points
y = 4 + 4 * X[:, 0] + 0.5 * X[:, 1] + np.random.randn(100)  # Generate y values with noise

# Add a column of ones to X for the bias terms (intercept) in the normal equation
# Use np.c_ to add a column on the left
X_bias = np.c_[np.ones(X.shape[0]), X]
```

2. **Normal Equation:**
  To solve for the parameters (intercept and coefficients) of the linear model, the code uses the normal equation method. The normal equation for linear regression is:

    $$ \theta = (X^T X)^{-1} X^T y $$

  where:
    - $X$ is the matrix of input data (with an added column of ones for the bias term).
    - $y$ is the vector of output values.
    - $ \theta $ is the vector containing the intercept and the coefficients.

  By computing the values of $ \theta $, the code retrieves the intercept (bias) and the coefficients (weights) for the linear model.

3. **Model Results:**
  After computing $ \theta $, the intercept and coefficients are printed. The intercept represents the value of $y$ when all input features $x_1$ and $x_2$ are zero, and the coefficients represent the change in $y$ with respect to a unit change in each of the input features.

```python
# Find theta using the normal equation X^TX theta = X^Ty
theta = np.linalg.inv(X_bias.T @ X_bias) @ X_bias.T @ y

# Print the intercept and coefficients of the regression plane
intercept = theta[0]
coefficients = theta[1:]

print(f"The intercept (bias) is: {intercept:.4f}")
print(f"The coeficients (weight) are: {coefficients}")
```

4. **Visualization:**
We generate a 3D scatter plot of the data points and overlays the regression plane. The regression plane is calculated predicting the $y$ values for the $ x_1 $ and $x_2 $ values using the fitted coefficients. The surface plot of the plane is displayed alongside the data points to visually show how well the regression model fits the data.

```python
import matplotlib.pyplot as plt
# Plot the data and regression plane
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of the data points
ax.scatter(X[:, 0], X[:, 1], y, color='red', label='Original Data')

# Create a grid for plotting the regression plane
x1 = np.linspace(X[:, 0].min(), X[:, 0].max(), 20)
x2 = np.linspace(X[:, 1].min(), X[:, 1].max(), 20)
x1_grid, x2_grid = np.meshgrid(x1, x2)
y_pred = intercept + coefficients[0] * x1_grid + coefficients[1] * x2_grid

# Plot the regression plane
ax.plot_surface(x1_grid, x2_grid, y_pred, cmap='viridis', alpha=0.8, label='Regression Plane')

# Labels and legend
ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("y")
plt.title("Linear Regression Plane")
plt.legend()
plt.show()
```

# **Fine-tuning the learning_rate and num_epochs**

To determine the feasible value of the learning rate and the number of epochs, we will use an empirical technique, which consists of iterating over a series of values ​​and identifying which combination produces the best results.

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    """Compute the sigmoid activation function."""
    return 1 / (1 + np.exp(-z))

# Set a random seed for reproducibility
np.random.seed(595)

# Generate 2D synthetic data points
X = np.random.rand(2, 100)  # Two features, 100 samples
w_true = np.array([1.5, -2.5])  # True weights for synthetic data
b_true = 1.0  # True bias for synthetic data

# Compute probabilities and assign binary labels
probabilities = sigmoid(np.dot(w_true.T, X) + b_true)
Y = (probabilities > 0.5).astype(int)

# Add noise to the features
X = X + 0.3 * np.random.rand(2, 100) - 0.1 * np.random.rand(2, 100)

# Split data into training and testing sets
X_train, X_test = X[:, :80], X[:, 80:]
Y_train, Y_test = Y[:80], Y[80:]

# Initialize tracking variables for optimal parameters
optimal_learning_rate = None
optimal_num_epochs = None

# Loop through a range of learning rates and epochs
learning_rates = [0.01,0.10,0.20,0.3,0.4,0.5,0.6,0.7,0.8,0.9,1.00]
epoch_values =   [100,200,300,400,500]

for learning_rate in learning_rates:
    for num_epochs in epoch_values:
        # Initialize model parameters
        w = np.zeros(X.shape[0])
        b = 0.0

        # Gradient Descent
        for epoch in range(num_epochs):
            # Forward pass
            A_train = sigmoid(np.dot(w.T, X_train) + b)

            # Compute gradients
            dJdw = np.dot(X_train, (A_train - Y_train).T) / len(Y_train)
            dJdb = np.mean(A_train - Y_train)

            # Update weights and bias
            w -= learning_rate * dJdw
            b -= learning_rate * dJdb

        # Evaluate accuracy on training set
        A_train = sigmoid(np.dot(w.T, X_train) + b)
        predictions_train = (A_train > 0.5).astype(int)
        train_accuracy = np.mean(predictions_train == Y_train)

        # Evaluate accuracy on testing set
        A_test = sigmoid(np.dot(w.T, X_test) + b)
        predictions_test = (A_test > 0.5).astype(int)
        test_accuracy = np.mean(predictions_test == Y_test)

        # Print progress
        print(f"Learning Rate: {learning_rate}, Epochs: {num_epochs}, "
              f"Train Accuracy: {train_accuracy:.2f}, Test Accuracy: {test_accuracy:.2f}")

        # Check stopping condition
        if train_accuracy >= 0.88 and test_accuracy >= 0.95:
            optimal_learning_rate = learning_rate
            optimal_num_epochs = num_epochs
            print("Stopping criterion met!")
            break  # Exit epoch loop
    if optimal_learning_rate is not None:
        break  # Exit learning rate loop

# Final output
if optimal_learning_rate is not None:
    print(f"\nOptimal Learning Rate: {optimal_learning_rate}")
    print(f"Optimal Number of Epochs: {optimal_num_epochs}")
else:
    print("\nNo combination met the stopping criterion.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    """Compute the sigmoid activation function.

    Args:
        z (array-like): Linear combination of weights, inputs, and bias.

    Returns:
        array-like: Sigmoid-transformed output values.
    """
    return 1 / (1 + np.exp(-z))

# Set a random seed for reproducibility
np.random.seed(595)

# Generate 2D synthetic data points
X = np.random.rand(2, 100)      # 2 features, 100 samples
w_true = np.array([1.5, -2.5])  # True weights for synthetic data
b_true = 1.0  # True bias for synthetic data

# Compute probabilities using the sigmoid function
probabilities = sigmoid(np.dot(w_true.T, X) + b_true)

# Assign binary labels (0 or 1) based on probabilities
Y = (probabilities > 0.5).astype(int)

# Add noise to the features to make the problem more realistic
X = X + 0.3 * np.random.rand(2, 100) - 0.1 * np.random.rand(2, 100)

# Split data into training (80%) and testing (20%) sets
X_train, X_test = X[:, :80], X[:, 80:]  # 80 samples for training, 20 for testing
Y_train, Y_test = Y[:80], Y[80:]        # Corresponding labels

# Initialize model parameters (weights and bias)
w = np.zeros(X.shape[0])  # Initialize weights to zeros
b = 0.0                   # Initialize bias to zero

# Hyperparameters
learning_rate = 0.5 # Learning rate for gradient descent
num_epochs = 500    # Number of iterations for training

# Gradient Descent Optimization
for epoch in range(num_epochs):
    # Forward pass: Compute predictions using sigmoid
    A_train = sigmoid(np.dot(w.T, X_train) + b)

    # Compute gradients of the cost function with respect to weights and bias
    dJdw = np.dot(X_train, (A_train - Y_train).T) / len(Y_train)  # Gradient for weights
    dJdb = np.mean(A_train - Y_train)                             # Gradient for bias

    # Update weights and bias using gradient descent
    w -= learning_rate * dJdw
    b -= learning_rate * dJdb

# Evaluate performance on training data
A_train = sigmoid(np.dot(w.T, X_train) + b)             # Compute predictions
predictions_train = (A_train > 0.5).astype(int)         # Threshold predictions
train_accuracy = np.mean(predictions_train == Y_train)  # Accuracy

# Evaluate performance on testing data
A_test = sigmoid(np.dot(w.T, X_test) + b)             # Compute predictions
predictions_test = (A_test > 0.5).astype(int)         # Threshold predictions
test_accuracy = np.mean(predictions_test == Y_test)   # Accuracy

# Print accuracy results
print(f"Training Set Accuracy: {train_accuracy:.2f}")
print(f"Test Set Accuracy: {test_accuracy:.2f}")
```
